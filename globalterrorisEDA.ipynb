# *BEGINS*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Global Terrorism Data.csv', encoding='latin1')
df.head()
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
df.fillna(0, inplace= True)
```

```python
df['iyear'].value_counts().sort_index().plot(kind='line')
plt.title("Attacks per Year")
plt.xlabel("Year")
plt.ylabel("Attacks")
plt.show()
```

```python
df['country_txt'].value_counts().head(20).plot(kind='bar')
plt.title("Top 20 Counrtries Attacks per Country")
plt.xlabel("Country")
plt.ylabel("Attacks")
plt.show()
```

```python
df['weaptype1_txt'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Weapon Types")
plt.xlabel("Weapon Type")
plt.ylabel("Attacks")
plt.show()
```

```python
df.groupby('iyear')['nkill'].sum().plot(kind="line")
plt.title("Total Killed per Year")
#plt.xlabel("Year")
#plt.ylabel("Killed")
plt.show()
```
